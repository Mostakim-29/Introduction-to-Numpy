```python
monthly_industry_sales = monthly_sales.sum(axis=1, keepdims=True)
print(monthly_industry_sales)
```

```python
monthly_sales_with_total = np.concatenate((monthly_sales, monthly_industry_sales), axis = 1)
print(monthly_sales_with_total)
```

```python
avg_monthly_sales = monthly_sales.mean(axis = 1)
print(avg_monthly_sales)
```

```python
plt.plot(np.arange(1,13), avg_monthly_sales,
         label="Average sales across industries")
plt.legend()
plt.show()
```

```python
plt.plot(np.arange(1,13), monthly_sales[:,2],
         label="Department store sales")
plt.legend()
plt.show()
```

```python
cumulative_monthly_industry_sales = monthly_sales.cumsum(axis = 0)
print(cumulative_monthly_industry_sales)
```

```python
plt.plot(np.arange(1,13), cumulative_monthly_industry_sales[:,0],
         label='Liquor Stores')
plt.plot(np.arange(1,13), cumulative_monthly_industry_sales[:,1],
         label='Restaurants')
plt.plot(np.arange(1,13), cumulative_monthly_industry_sales[:,2],
         label='Department stores')
plt.legend()
plt.show()
```
